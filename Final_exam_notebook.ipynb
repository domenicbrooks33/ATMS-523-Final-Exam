```python
import pandas as pd
import datetime
```

# First, read in station data

```python
# Station metadata
stn_ids = pd.read_fwf('http://noaa-ghcn-pds.s3.amazonaws.com/ghcnd-stations.txt', header=None, infer_nrows=1000)
stn_ids.columns = ['ID','LAT','LON','ELEV','UKN','NAME','GSN','WBAN']
stn_ids
```

```python
# Station periods of record
periods = pd.read_fwf('http://noaa-ghcn-pds.s3.amazonaws.com/ghcnd-inventory.txt', header=None, infer_nrows=1000)
periods.columns = ['ID','LAT','LON','ELEM','TiMIN','TiMAX']
periods
```

```python
# Merge the two DataFrames
merged_stns = pd.merge(stn_ids,periods,how='left',left_on='ID',right_on='ID')
merged_stns
```

```python
# Grab all stations that have TMIN data from at least 1990 to 2020
merged_stns = merged_stns[(merged_stns['ELEM'] == 'TMIN') & (merged_stns['TiMAX'] >= 2020) & (merged_stns['TiMIN'] <= 1990)]
merged_stns
```

```python
station_id = 'USC00087205'
# Make sure our station exists and its period of record is sufficient
merged_stns.loc[merged_stns['ID'] == station_id]
```

```python
# Load the station data from S3 bucket
df = pd.read_csv(
    f"s3://noaa-ghcn-pds/csv/by_station/{station_id}.csv",
    storage_options={"anon": True},  # passed to `s3fs.S3FileSystem`
    dtype={'Q_FLAG': 'object', 'M_FLAG': 'object'},
    parse_dates=['DATE']
).set_index('DATE')

df
```

```python
# Grab TMIN data since we are concerned with freezing days
df_tmin = df.loc[df['ELEMENT'] == 'TMIN']
df_tmin['DATA_VALUE'] = df_tmin['DATA_VALUE'] / 10.0  # Convert to deg C
df_tmin
```

### Now that we have the min temp data from Plant City, lets move onto question 1.

Goal: Mean number of days per month (for October-January) for 1991-2020 period that were less than or equal to 32 deg F (frost) and 28 deg F (freeze).

```python
# Convert to deg F first
df_tmin_f = df_tmin.copy()
df_tmin_f['DATA_VALUE'] = ((df_tmin['DATA_VALUE'] * (9/5)) + 32) 
df_tmin_f
```

```python
# Select TMIN data for 1991-2020 and months October to January
df_tmin_f_period = df_tmin_f[(df_tmin_f.index.year >= 1991)
                              & (df_tmin_f.index.year <= 2020)
                              & (df_tmin_f.index.month.isin([10,11,12,1]))]

# Add month of year column as well
df_tmin_f_period['MONTH'] = df_tmin_f_period.index.month
df_tmin_f_period
```

```python
# Add columns to denote frost and freeze days
df_tmin_f_period['FROST'] = df_tmin_f_period['DATA_VALUE'] <= 32
df_tmin_f_period['FREEZE'] = df_tmin_f_period['DATA_VALUE'] <= 28

df_tmin_f_period
```

```python
# Number of years in the period
num_years = len(df_tmin_f_period.index.year.unique().values)
```

```python
# Group by month and sum the number of frost and freeze days over the 30-year period
df_mean_ff_days = df_tmin_f_period.groupby('MONTH').agg({'FROST':'sum','FREEZE':'sum'}) / num_years  # Divide by 30 years to get mean number of days per month
df_mean_ff_days
```

```python
print('Mean Number of Frost Days:\n'
      f'Jan: {df_mean_ff_days["FROST"].loc[1]:.2f}\n'
      f'Oct: {df_mean_ff_days["FROST"].loc[10]:.2f}\n'
      f'Nov: {df_mean_ff_days["FROST"].loc[11]:.2f}\n'
      f'Dec: {df_mean_ff_days["FROST"].loc[12]:.2f}\n')

print('Mean Number of Freeze Days:\n'
      f'Jan: {df_mean_ff_days["FREEZE"].loc[1]:.2f}\n'
      f'Oct: {df_mean_ff_days["FREEZE"].loc[10]:.2f}\n'
      f'Nov: {df_mean_ff_days["FREEZE"].loc[11]:.2f}\n'
      f'Dec: {df_mean_ff_days["FREEZE"].loc[12]:.2f}\n')
```

### Part 2: ENSO patterns

Goal: Determine which ENSO region is best correlated (highest Pearson-correlation) with the number of days per month <28 def F (Freeze days)

```python
# Read in full ENSO dataset
full_enso = pd.read_csv('https://www.cpc.ncep.noaa.gov/data/indices/sstoi.indices',
                        sep=r'\s+', header=0)

# Rename month column for merging later
full_enso.rename(columns={'MON': 'MONTH'}, inplace=True)
full_enso.sort_values(['YR','MONTH'], inplace=True) # Sort just to be sure
full_enso
```

```python
# Creat a new DataFrame with a Date index covering the full period of the ENSO dataset
enso_new=pd.DataFrame()
enso_new['Date']=pd.date_range(start=datetime.datetime(full_enso['YR'][0],1,1),end=datetime.datetime(full_enso['YR'].iloc[-1],12,1),freq="MS")

# Set Date as index and add MONTH column
enso_new = enso_new.set_index('Date').sort_index()
enso_new['MONTH'] = enso_new.index.month
enso_new = enso_new.iloc[:-1] # Drop last row since it goes beyond the ENSO data period

# Add ENSO anomaly columns to new DataFrame
enso_new['anom1+2'] = full_enso['ANOM'].values
enso_new['anom3'] = full_enso['ANOM.1'].values
enso_new['anom4'] = full_enso['ANOM.2'].values
enso_new['anom3.4'] = full_enso['ANOM.3'].values
enso_new
```

```python
# First need to sum up freeze days by month and year
df_tmin_f_period['FREEZE_days_alt'] = df_tmin_f_period['DATA_VALUE'] < 28 # Only days <28F 
df_monthly_freeze_days = df_tmin_f_period.groupby(pd.Grouper(freq='MS')).agg({'FREEZE_days_alt':'sum'})
df_monthly_freeze_days
```

```python
# Now merge the two DataFrames on shared date index
df_merged = df_monthly_freeze_days.merge(enso_new, left_index=True, right_index=True, how='left')
df_merged = df_merged[df_merged.index.month.isin([10,11,12,1])] # Keep only Oct-Jan months
df_merged
```

```python
# Correlation between freeze days and each anomaly
print('Pearson correlations:\n')
for col in ['anom1+2', 'anom3', 'anom4', 'anom3.4']:
    corr = df_merged['FREEZE_days_alt'].corr(df_merged[col], method='pearson')
    print(f'{col}: {corr:.4f}')
```

### It would appear as though all the ENSO regions are weakly negatively correlated with the monthly number of days <28 deg F for Oct-Jan, but the region with the highest magnitude correlation appears to be region 1+2, with -0.1356.
